## 1.1 Descri√ß√£o e motiva√ß√£o do problema


<p>O suic√≠dio √© um fen√¥meno social que afeta todos os pa√≠ses do mundo, constituindo um grave problema de sa√∫de p√∫blica. Cerca de 700 mil indiv√≠duos morrem todos os anos, sendo principalmente jovens e adultos de meia-idade que tentam o suic√≠dio (MARTINS, 2021). No Brasil, √© a terceira causa de morte de homens e a quarta causa de morte de mulheres entre os 15 e 29 anos de idade (MALTA, 2021).A ONU inseriu como um de seus objetivos de desenvolvimento sustent√°vel a redu√ß√£o da taxa de suic√≠dio global registrada 2013 um ter√ßo at√© 2030, enquanto e a OMS estabeleceu como meta 15% de redu√ß√£o da mesma taxa registrada em 2019 at√© 2023, com base em valores de 2019. (WHO, 2021)</p><br>
<center><figure><figure><img src="https://jupsin.com/wp-content/uploads/2021/06/1Captura-de-pantalla-2021-06-22-a-las-12_Fotor-1-1024x549.jpeg"><br><figcaption>Pilares do guia de preven√ß√£o ao suic√≠dio da OMS<figcaption> </figure></center>

<p> A idea√ß√£o suicida √© um problema que afeta indiv√≠duos de todas as idades em todo o mundo devido a diversos fatores, como desgosto, raiva, culpa e outros sintomas de melancolia ou ansiedade. A idea√ß√£o suicida √© um estado em que uma pessoa tem pensamentos recorrentes associados ao desejo de morrer, seja por se considerar um fardo para sua fam√≠lia ou ambiente, seja por ter um sentimento de n√£o pertencimento ao grupo ou sociedade a que pertence. Embora a maioria das pessoas com pensamentos suicidas (80%) n√£o tente o suic√≠dio, a depress√£o pode levar ao suic√≠dio se n√£o for tratada de maneira eficaz a longo prazo (DENIS-RODRIGUEZ, 2017).</p>

<p>Um dos pilares do guia de preven√ß√£o ao suic√≠dio elaborado pela OMS √© ‚Äúa identifica√ß√£o precoce, avalia√ß√£o, gerenciamento e acompanhamento de pessoas acometidas por comportamentos suicidas‚Äù(WHO, 2021).A detec√ß√£o precoce de idea√ß√µes suicidas em postagens pode ent√£o ajudar os especialistas em sa√∫de a identificar indiv√≠duos em risco. Assim, um sistema automatizado de detec√ß√£o de idea√ß√£o suicida e primeiros sintomas de depress√£o em postagens online permitiria a estes profissionais fornecer o tratamento mais adequado e poupar muitas vidas.</p>
<p>An√°lise de Sentimento, ou minera√ß√£o de opini√£o, √© um campo crescente que captura automaticamente o sentimento do usu√°rio (SILVA, 2018). A combina√ß√£o de informa√ß√µes de m√≠dia social e An√°lise de Sentimento pode reconhecer pensamentos suicidas precoces e evitar a maioria das tentativas de suic√≠dio. Como resultado, <i>Machine Learning</i> e <i>Natural Language Processing</i> passaram a ser aplicadas como t√©cnicas para prever a inten√ß√£o suicida a partir de dados de m√≠dia social. Al√©m disso, as arquiteturas de <i>Deep Learning</i> fornecem vantagens significativas para detectar pensamentos suicidas, uma vez que estes m√©todos executam com precis√£o muito alta com n√≠vel relativamente baixo de processamento (M√ÑNTYL√Ñ, 2018).</p>
<p>Estudos anteriores que usaram algoritmos de <i>Machine Learning</i> para identificar idea√ß√µes suicidas em postagens foram limitados em tamanho de conjunto de dados. A pesquisa de Pachouly (2021) usou v√°rios modelos de <i>Machine Learning</i> para detectar depress√£o em uma cole√ß√£o de 15.000 tweets, mas a baixa precis√£o foi um problema elencado. Um trabalho semelhante foi realizado por Manisha (2019), que melhorou o desempenho dos classificadores de <i>Machine Learning</i> em um conjunto de dados de 50.000 tweets coletados manualmente e rotulados para classifica√ß√£o bin√°ria. Em Stankevich (2019), os autores propuseram um sistema autom√°tico de detec√ß√£o de depress√£o usando modelos de <i>Machine Learning</i> em um conjunto de dados criado a partir de uma rede social russa, Vkontakte.</p>
<p>No entanto, esses estudos n√£o obtiveram uma boa precis√£o devido ao treinamento de algoritmos de <i>Machine Learning</i> em uma pequena cole√ß√£o de texto. Com o uso de regras de anota√ß√£o apropriadas, √© poss√≠vel criar uma base de dados maior para o treinamento de modelos de <i>Deep Learning</i> e melhorar a classifica√ß√£o de modelos de <i>Machine Learning</i>. Estudos recentes listados por Abdulsalam (2022) mostram que classificadores <i>Deeping Learning</i> foram usados para identificar inten√ß√µes suicidas de usu√°rios de redes sociais com maior precis√£o, principalmente em conjuntos de dados anotados manualmente coletados de f√≥runs de depress√£o e suic√≠dio (ALADAG, 2018) e <i>subreddits</i> (SHAH, 2020).</p>
<p>Neste trabalho, seguiremos essa abordagem e iremos implementar um <i>Deeping Learning</i> na tarefa de identificar idea√ß√µes suicidas em postagens de f√≥runs de redes sociais. A principal contribui√ß√£o deste estudo √© avaliar experimentalmente o desempenho de um classificador de Deep Learning em um conjunto de dados de aproximadamente 232.000 postagens, de f√≥runs de uma rede social, rotuladas como "suic√≠dio" e "n√£o-suic√≠dio".</p>

### 2.3.1 - Stop Words e Normaliza√ß√£o

```python
nltk.download('stopwords')
```

```python
stop_words = stopwords.words("english")
stemmer = SnowballStemmer("english")
```

```python
def preprocess(text, stem=False):
    # Remover link,user e caracteres especiais
    text = re.sub("@\S+|https?:\S+|http?:\S|[^A-Za-z0-9]+", ' ', str(text).lower()).strip()
    tokens = []
    for token in text.split():
        if token not in stop_words:
            if stem:
                tokens.append(stemmer.stem(token))
            else:
                tokens.append(token)
    return " ".join(tokens)
```

```python
%%time
df_raw['text'] = df_raw['text'].apply(lambda x: preprocess(x))
```

### 2.3.2 Label Encoder
<p> Ser√° aplicado o label encoder na coluna da classifica√ß√£o do texto

###  2.3.2 Word2Vec
<p>Caracteriza√ß√£o de cada palavra, pelas outras palavras em seu entorno.</p>

```python
%%time
documents = [_text.split() for _text in df_train['text'].astype(str)]
```

```python
#Carrega o w2v, n√£o existir
if (os.path.exists('./models/w2v_model.w2v')):
    w2v_model = Word2Vec.load('./models/w2v_model.w2v')
else:
    w2v_model = Word2Vec(sentences=documents, vector_size=300, window=8,workers=8, min_count=10, epochs=32)
    w2v_model.train(documents, total_examples=w2v_model.corpus_count, epochs=w2v_model.epochs)
```

```python
words = w2v_model.wv.key_to_index.keys()
vocab_size = len(words)
print(f"Tamanho do vocabul√°rio {vocab_size}, corpus {w2v_model.corpus_count}")
```
